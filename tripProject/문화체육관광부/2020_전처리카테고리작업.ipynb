```python
for index in df.index:
  if df.loc[index, 'sleepnights'] == '0':
    df.loc[index, 'accommodation'] = 0
  else:
    nights = list(map(int, df.loc[index, 'sleepnights'].split(',')))
    try:
      acco = list(map(int, df.loc[index, 'accommodation'].split(',')))
    except:
      for i in range(1,len(nights)+1):
        col_name = 'accommodation_'+str(i)
        df[col_name]=0
        pass
    if len(nights)==len(acco):
      pass
    elif sum(nights)==len(acco): ## ex [0,1] [1] -> [0,1]
      for no, n in enumerate(nights):
        if n == '0':
          acco.insert(no,0)
          try:
            for acco_i, acc in acco: #[0,0,0,1]
              col_name = 'accommodation_' + str(acco_i)
              df.loc[index,col_name] = acc
          except:print(index)
    else: # 3,3 일때 3,12 -> 3,3,3,12,12,12
      sample_list = []
      for no, night in enumerate(nights):
        if night == '0':
          acco.insert(no,0)
        for i in range(night):
          try:
            sample_list.append(acco[no])
          except:
            print(index)

      try:
        for sample_i, sample in sample_list: #[3,3,3, 12, 12, 12]
            col_name = 'accommodation_' + str(sample_i)
            df.loc[index,col_name] = sample
      except:print(index)
        


```
